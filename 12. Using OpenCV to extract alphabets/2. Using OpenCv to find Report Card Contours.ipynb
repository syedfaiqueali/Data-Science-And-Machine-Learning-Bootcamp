# Imports

```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image

%matplotlib inline
```

# Constants

```python
CROPPED_CONTOURS_FOLDER = 'Cropped_Contours/'
SAMPLE_IMAGE = 'Image_Samples/Sample1.png'
```

# Helper Functions

```python
def get_all_childs_of_parent(parent, hierarchy):
    '''
    This function will return all the child of that contour
    Parameters:
    parent : Int value
    hierarchy: contours hierarchy
    '''
    # Initialize
    hier_len = hierarchy.shape[1] #Total contours
    child_contours = {} #Empty Dict
    
    for x in range(hier_len):
        # To loop through whole hierarchy
        next_contour, pre_contour, first_child, parent_contour = hierarchy[0][x]
        
        # If desired parent found, so append it all childs
        if parent == parent_contour:
            child_contours[x] = hierarchy[0][x]
        
    # return all the found contours
    return child_contours


def remove_extraneous_contours(index_to_keep, contours_dict):
    '''
    This function will remove extraneous contours from all child contours
    Parameters:
    index_to_keep : Array containing hierarchy values to keep
    contours_dict: All child contours dict
    '''
    # Initialize an empty dict
    final_dict = {}
    
    # Loop through all contours child dict and search for index_to_keep keys 
    for x in index_to_keep:
        final_dict[x] = contours_dict[x]
          
    # Return only desired childs        
    return final_dict


def draw_bounding_box_on_desired_cells(desired_contours, contours):
    '''
    This function will draw bouding box on the cells
    Parameters:
    desired_contours : Cells on which want to draw the box
    contours: All contours hierarchy array
    '''
    # Make an original image copy
    image_new_copy_2 = img.copy()
    
    # Loop through every cell contour and draw bounding box
    for x in range(len(desired_contours)):
        cnt = contours[desired_contours[x]]
        x,y,w,h = cv2.boundingRect(cnt)
        plt.imshow(cv2.rectangle(image_new_copy_2, (x,y), (x+w,y+h), (225,0,0), 1))
       
    # To show the new image on which cells are drawn
    plt.show()
    

def draw_bounding_box_on_desired_cells_elements(desired_contours, contours, hierarchy):
    '''
    This function will draw bouding box inside cell elements
    Parameters:
    desired_contours : Cells on which want to draw the box on their ele
    contours: All contours array
    hierarchy: All contours hierarchy
    '''
    # Make an original image copy
    image_new_copy_2 = img.copy()
    # Initializations
    rect_color = (255,0,0)
    rect_stroke_width = 1
    
    # Loop through every cell contour and 
    # draw bounding box on every cell first childs
    for i in range(len(desired_contours)):
        childs = get_all_childs_of_parent(desired_contours[i],hierarchy)
        
        for child in childs:
            cnt = contours[child]
            x,y,w,h = cv2.boundingRect(cnt)
            plt.imshow(cv2.rectangle(image_new_copy_2, (x,y), (x+w,y+h), rect_color, rect_stroke_width))
            
            # Cropping individual contours and saving them as separate image
            cropping_rectangles_and_saving_them(x,y,x+w,y+h,SAMPLE_IMAGE,child)
          
    # To show the new image on which cells elements are bounded 
    plt.show()
    
    
def cropping_rectangles_and_saving_them(left,upper,right,lower, image_name, cropped_image_name):
    '''
    This function will crop all the bounding rectangles from an image
    Parameters:
    left=x,upper=y,right=x+w,lower=y+h
    image_name: original image
    cropped_image_name: individual cropped images name
    '''
    # Opens image using PIL
    im = Image.open(image_name)
    
    # Crop image from original image
    box = (left,upper,right,lower)
    cropped_image = im.crop(box)
    
    # Saved cropped image
    cropped_image.save(f'{CROPPED_CONTOURS_FOLDER}{cropped_image_name}.png')
    
    
def draw_rectangle_on_img_contours(contours, image):
    '''
    This function will draw rectangles on images
    Parameters:
    contours : array
    image: image
    '''
    total_contours = len(contours)
    rect_color = (255,0,0)
    rect_stroke_width = 1
    
    for i in range(total_contours):
        cnt = contours[i]
        x,y,w,h = cv2.boundingRect(cnt)
        image = cv2.rectangle(image, (x,y), (x+w,y+h), rect_color, rect_stroke_width)
        
        # Cropping individual contours and saving them as separate image
        cropping_rectangles_and_saving_them(x,y,x+w,y+h,SAMPLE_IMAGE,i)
        
    # To visualize all contours on image
    plt.imshow(image)
    plt.show()
```

# Image Processing

### 1. Converting Image to B&W

```python
# Read image using opencv
img = cv2.imread(SAMPLE_IMAGE)

# Display image using matplotlib
plt.imshow(img)
plt.show()
#cv2.destroyAllWindows()
```

```python
# convert the image to grayscale format
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray)
plt.show()
```

```python
# apply binary thresholding
ret, thresh = cv2.threshold(img_gray, 150, 255, cv2.THRESH_BINARY)
#cv2.imwrite('image_thres1.jpg', thresh)

plt.imshow(thresh)
plt.show()
```

### 2. Find image contours

```python
# detect the contours on the binary image using cv2.CHAIN_APPROX_NONE
contours, hierarchy = cv2.findContours(image=thresh, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_NONE)
                                     
# draw contours on the original image
image_copy = img.copy()
cv2.drawContours(image=image_copy, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2, lineType=cv2.LINE_AA)
               
plt.imshow(image_copy)
plt.show()

#cv2.imwrite('contours_none_image1.jpg', image_copy)
```
